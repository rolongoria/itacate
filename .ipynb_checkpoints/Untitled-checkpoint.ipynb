```python
# Funcion para imprimir la cantidad vendida por parentMenuSelection
def sumaProductosParentSelection(parentMenuSelection):
    print(parentMenuSelection)
    
    modifierArray = df2[df2['Parent Menu Selection'] == parentMenuSelection]['Modifier'].unique()
    
    for modifier in modifierArray:
        print(modifier + ' ' + str(df2[(df2['Parent Menu Selection'] == parentMenuSelection) & (df2['Modifier'] == modifier)]['Qty'].sum()))
```

```python
def showByParentSelection(parentMenuSelection):
    return df2[df2['Parent Menu Selection'] == parentMenuSelection]
```

```python
def showModifiers(parentMenuSelection):
    return df2[df2['Parent Menu Selection'] == parentMenuSelection]['Modifier'].unique()
```
